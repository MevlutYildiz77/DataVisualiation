To use Matplotlib plots on python

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
matplotlib.pyplot
```

```python
x =[1,2,3,4,5,6,7]
y = 50,51,52,48,47,49,46
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'g')
```

```python
plt.plot(x,y, 'g+')
```

```python
plt.plot(x,y, 'g+--')
```

```python
plt.plot(x,y, 'g+-.')
```

```python
plt.plot(x,y, '--r+')
```

```python
plt.plot(x,y, 'r+--')
```

```python
plt.plot(x,y, 'rD--')
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='#008000',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='#008040',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='',markersize=20)
```

```python
plt.plot(x,y,color='red',marker='+',linestyle='',markersize=20)
```

```python
# alpha used for the transperency of the plot. it is around 0-1
plt.plot(x,y,color='green',alpha=0.8)
```

```python
plt.plot(x,y,color='green',alpha=0.2)
```

```python
#Functional Method
plt.subplot(2,1,1)
plt.plot(x1, y1)
plt.subplot(2,1,2)
plt.plot(x2, x2)
```

```python
#Object-Oritented Method
fig,axes=plt.subplots(1,2)      # or
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x ,y)
```
